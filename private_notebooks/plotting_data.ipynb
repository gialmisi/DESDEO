```python
import plotly.graph_objects as go
import polars as pl

baseline_file = "new_branin_function_baseline_2025-11-26T10:38:37.137543_stats_0_18_100_500_c_1_f_1.parquet"
relaxed_file = "new_branin_function_relaxed_2025-11-26T11:31:50.272341_stats_6_18_100_500_c_1_f_1.parquet"

files = [baseline_file, relaxed_file]


def add_fig(fig, data, name):
    fig.add_trace(go.Scatter(x=data["generation"], y=data["best_mean"], mode="lines", name=name))
    fig.add_traces(
        go.Scatter(
            x=data["generation"].to_list() + data["generation"].to_list()[::-1],
            y=data["best_upper"].to_list() + data["best_lower"].to_list()[::-1],
            fill="toself",
            name="std one dev",
            line={"width": 0},
            # fillcolor="rgba(0,0,255,0.15)",
            opacity=0.25,
            hoverinfo="skip",
            showlegend=False,
        )
    )


fig = go.Figure()

for i, f_name in enumerate(files):
    data = pl.read_parquet("./data/" + f_name)
    add_fig(fig, data, f"Data {i}")


fig.update_layout(
    title=f"Best mean f_1 per generation (Relaxed vs Baseline)",
    xaxis_title="Generation",
    yaxis_title="Cumulative best mean f_1",
)

fig.show()
```
